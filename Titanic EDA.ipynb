```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_df = pd.read_csv('titanic.csv')
# View the first few rows of the dataset
titanic_df.head()
```

```python
# View the shape of the dataset
print(titanic_df.shape)
```

```python
# Check for missing values
print(titanic_df.isnull().sum())
```

```python
mean_age = titanic_df['Age'].mean()
titanic_df['Age'].fillna(mean_age, inplace=True)

# Replace missing values in the Cabin column with "Unknown"
titanic_df['Cabin'].fillna('Unknown', inplace=True)

# Replace missing values in the Embarked column with the mode (most common value)
mode_embarked = titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'].fillna(mode_embarked, inplace=True)

mean_fare = titanic_df['Fare'].mean()
titanic_df['Fare'].fillna(mean_fare, inplace=True)

# Check for missing values again
print(titanic_df.isnull().sum())
```

```python
# Plot the survival rate by sex
sns.barplot(x="Sex", y="Survived", data=titanic_df)
```

```python
# Plot the survival rate by passenger class
sns.barplot(x="Pclass", y="Survived", data=titanic_df)
```

```python
# Plot the distribution of ages
sns.histplot(data=titanic_df, x="Age", kde=True)

```

```python
# Plot a pairplot of the dataset
sns.pairplot(titanic_df, hue="Survived")
```
